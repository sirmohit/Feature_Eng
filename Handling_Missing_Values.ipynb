## Imputation techniques
1 -Mean Value imputation

```python
sns.distplot(df["age"])
```

```python
df.age.isnull().sum()
```

```python
# handle missing value when data is normaly distributed
df["Age_mean"] = df['age'].fillna(df['age'].mean())
```

```python
df[["Age_mean",'age']]
```

##Median Value Imputation

```python
# theis technique is usefull when data is skewed or outliers in dataset
df["Age_median"] = df['age'].fillna(df['age'].median())
```

```python
df[["Age_median",'age']]
```

# Mode value imputation

```python
# this technique is useful when missing value is categorical data
df['embarked'].isnull()
```

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
df[df['age'].notna()]['embarked'].mode()[0]
```

```python
mode = df[df['age'].notna()]['embarked'].mode()[0]
```

```python
df["embarked_mode"] = df['embarked'].fillna(mode)
```

```python
df[["embarked_mode",'embarked']]
```

```python
df["embarked"].isnull().sum()
```
